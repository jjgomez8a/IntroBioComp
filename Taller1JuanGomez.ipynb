```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos inventados, tambien se puso una tercera variable
data = {
    'Familia': ['Apidae', 'Halictidae', 'Megachilidae', 'Andrenidae',
            'Colletidae', 'Melittidae', 'Stenotritidae', 'Nomadinae',
            'Xylocopinae', 'Bombini', 'Euglossini', 'Ctenoplectrini'],
    'poblacion': [3203, 150, 945, 72, 524, 433, 38, 25, 1469, 142, 3441, 94],
    'NectarProducido' : [1780, 950, 1120, 1340, 890, 1230, 1675, 740, 1520, 990, 1105, 1420]
}

# Convertimos el diccionario en un DataFrame
df = pd.DataFrame(data)

# Calcular eficiencia de producción de néctar por abeja (%)
df['Eficiencia'] = df['NectarProducido'] / df['poblacion'] * 100

# Ordenamos el DataFrame de mayor a menor segun la eficiencia de las abejas (%)
df.sort_values(by='Eficiencia', ascending=False, inplace=True)
# Mostramos la tabla ordenada
print(df)
```

```python
plt.figure(figsize=(10,6))
plt.barh(df['Familia'], df['Eficiencia'],
         # variacion de color en tabla
         color=['yellow' if i % 2 == 0 else 'black' for i in range(len(df))])
 # titulos
plt.xlabel('ml por abeja')
plt.title('Eficiencia por individuo')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
