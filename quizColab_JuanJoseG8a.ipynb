```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
'Especie': ["Lungfish","Shark","Frog","Salamander","Turtle","Alligator","Falcon","Ostrich","Cod","Armadillo","Elephant","Human","Dog","Wombat","Platypus","Amphioxus"],
'A': [301, 311,336,319,361,333,340,307,288,359,352,326,331,340,309,243],
'C':[338,293,289,319,359,366,382,387,307,328,326,389,318,366,315,211],
'G':[152,150,144,166,131,145,142,147,182,150,137,136,159,144,157,246],
'T':[353,390,371,337,293,315,279,299,384,303,322,286,332,296,359,443]
}
# Convertimos el diccionario en un DataFrame (tabla de datos) usando pandas
df = pd.DataFrame(data)

df["%GC"] = ((df["G"] + df["C"]) / (df["A"] + df["C"] + df["G"] + df["T"])) * 100

df = df.sort_values(by="%GC", ascending=False)

# Graficar
plt.figure(figsize=(10,6))
plt.bar(df["Especie"], df["%GC"], color='green')
plt.xticks(rotation=90)
plt.ylabel("%GC")
plt.title("Contenido GC por Especie (ordenado de mayor a menor)")
plt.tight_layout()
plt.show()
print(df[["Especie", "%GC"]])

```

El análisis muestra que el contenido GC varía entre especies, con avestruz presentando el valor más alto (46.84%) y rana el más bajo (37.98%). En general, aves como el avestruz y halcon tienen porcentajes más elevados, mientras que anfibios como rana y peces como Cod y tiburon presentan valores menores. Aun asi, el elefante aparece con un valor relativamente bajo (40.72%), incluso por debajo de peces como el Cod (42.12%).


---


Esto indica que el contenido GC no sigue un patrón fijo por grupo taxonómico, más bien está influenciado por factores evolutivos específicos de cada linaje.
