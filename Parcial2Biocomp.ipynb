```python
#Cual es la longitud del gen en pares de bases
!pip -q install biopython
from Bio import SeqIO
ruta = "Carpeta_Examen/secuencia.fasta"
record = SeqIO.read(ruta, "fasta")
print("Longitud (bp):", len(record.seq))
##El print muestra 2517 pares de bases
```

```python
##Cual es su secuencia y longitud en amino ácidos?
from Bio import SeqIO
import os

ruta = "Carpeta_Examen/secuencia.fasta"
record = SeqIO.read(ruta, "fasta")
seq_aa = str(record.seq.translate(to_stop=True)).replace("*","")

salida = os.path.join("Carpeta_Examen",
                      os.path.splitext(os.path.basename(ruta))[0] + "_translated.faa")

with open(salida, "w") as f:
    f.write(f">{record.id}_translated\n")
    for i in range(0, len(seq_aa), 60):
        f.write(seq_aa[i:i+60] + "\n")

print("Longitud (aa):", len(seq_aa))
print("Proteína guardada en:", salida)
##El print muestra 838 AminoAcidos guardados en la carpeta secuencia_translated.faa
```
